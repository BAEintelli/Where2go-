### 동전한닢(MeCab) 사용한 토큰화

```python
from konlpy.tag import Mecab
```

```python
mecab = Mecab(dicpath="C:\\mecab\\mecab-ko-dic")
```

```python
df = pd.read_csv("./네이버 블로그 본문")
```

```python
test = df.iloc[0].description
```

```python
mecab.morphs(test)[:10]
```

```python
df["token_descirption"] = "1"
```

```python
df
```

```python
a = -10
b = 0

for i in range(99):
    a += 10
    b += 10
    test = df[a:b]
    for i, row in test.iterrows():
        contents_tokens = mecab.morphs(row.description)
        sentence = ''
        for word in contents_tokens:
            sentence = sentence+' '+word
        row["token_descirption"] = sentence
    test.to_csv("./mecab_token/{}-{}".format(a,b))
```

```python
from glob import glob

mecab_list = glob("./mecab_token/*") 
```

```python
#폴더에 있는 데이터프레임 하나로 합치기
startframe = mecab_list.pop(0)
final = pd.read_csv(startframe)
for frame in mecab_list:
    df = pd.read_csv(frame)
    final = pd.concat([final ,df])
```

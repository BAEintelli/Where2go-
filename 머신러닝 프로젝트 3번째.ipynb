### 네이버 블로그 본문 크롤링
- 블로그 본문을 크로링할때 메모리 이슈가 발생하여 데이터프레임에서 100row씩으로 끊어서 파일을 저장하였다

```python
a = -100
b = 0
for i in range(10):
    a += 100
    b += 100
    test = top10[a:b]
    dicts = {}
    for idx, row in test.iterrows():
        logno = row.link.split("logNo=")[1]
        iD = row.link.split("?")[0].split("com/")[1]
        url="https://blog.naver.com//PostView.nhn?blogId={}&logNo={}&from=search&redirect=Log&widgetTypeCall=true&topReferer=https%3A%2F%2Fsearch.naver.com&directAccess=true".format(iD, logno)
        respones = requests.get(url)
        dom = BeautifulSoup(respones.content, 'html.parser')
        try:
            if row.NAME not in dicts:
                dicts[row.NAME] = dom.select("#post-view{}".format(logno))[0].text
            else:
                dicts[row.NAME] += dom.select("#post-view{}".format(logno))[0].text
        except:
            pass
    
    test_ls = []
    test_ls.append(dicts)
    contents = pd.DataFrame(test_ls).T
    contents.columns = ["description"]
    contents.to_csv("./네이버블로그 크롤링/{}~{}.csv".format(a,b)) 
    
```
